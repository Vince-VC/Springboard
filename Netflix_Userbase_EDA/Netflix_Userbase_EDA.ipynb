```python
#import packages
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import scipy.stats as stats
import warnings

plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)

warnings.simplefilter(action="ignore", category=FutureWarning)
warnings.filterwarnings(action="ignore", module="scipy", message="^internal gelsd")
```

```python
#read in the dataset
df = pd.read_csv('Netflix_Userbase.csv')
```

```python
# Distribution of subscription type
type = df['Subscription Type']
type_hist = sns.histplot(data=type)
type_hist = plt.xlabel('Subscription Type')
type_hist = plt.ylabel('Count')
type_hist = plt.title('Subscription Type Distribution')
```

```python
# Distribution of Monthly Revenue 
revenue = df['Monthly Revenue']
revenue_hist = sns.histplot(data=revenue)
revenue_hist = plt.xlabel('Monthly Revenue')
revenue_hist = plt.ylabel('Count')
revenue_hist = plt.title('Monthly Revenue Distribution')
```

```python
# Distribution of country 
country = df['Country']
country_hist = sns.histplot(data=country)
country_hist = plt.xlabel('Country')
country_hist = plt.xticks(rotation=90)
country_hist = plt.ylabel('Count')
country_hist = plt.title('Country Distribution')
```

```python
# Distribution of Age
age = df['Age']
age_hist = sns.histplot(data=age, bins=26)
age_hist = plt.xlabel('Age')
age_hist = plt.ylabel('Count')
age_hist = plt.title('Age Distribution')
```

```python
# Distribution of gender
gender = df['Gender']
gender_hist = sns.histplot(data=gender)
gender_hist = plt.xlabel('Gender')
gender_hist = plt.ylabel('Count')
gender_hist = plt.title('Gender Distribution')
```

```python
# Distribution of device 
device = df['Device']
device_hist = sns.histplot(data=device)
device_hist = plt.xlabel('Device')
device_hist = plt.ylabel('Count')
device_hist = plt.title('Device Distribution')
```

```python
# Pointplot of subscription type v age per gender
age_hist1 = sns.pointplot(x='Subscription Type', y='Age', data=df, hue='Gender')
age_hist1 = plt.xlabel('Subscription Type')
age_hist1 = plt.ylabel('Age')
age_hist1 = plt.title('Catplot of Subscription Type v Age per Gender')
```
