**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
#Calculate the average price per borough in 1998 and in 2018 and reset index
pd.options.display.float_format = '{:.2f}'.format
Avg_Price_per_Year = properties_98_18.groupby(["Borough Name", "Year"])["Average Price"].mean()
Avg_Price_per_Year = Avg_Price_per_Year.reset_index()
pd.set_option('display.max_rows', None)
Avg_PPY_DF = pd.DataFrame(Avg_Price_per_Year)
Avg_PPY_DF.head()
```

```python
#Calculate the difference between each borough's 2018 value and 1998 value to find the total price increase over that time and create a new column with the result
Avg_PPY_DF["Total Price Increase"] = Avg_PPY_DF.groupby("Borough Name")["Average Price"].diff().fillna(0)
Avg_PPY_DF.head()
```

```python
#Drop 1998 data 
Total_Increase_DF = Avg_PPY_DF.loc[Avg_PPY_DF['Year'] != 1998]
Total_Increase_DF.head()
```

```python
#Bar plot showing the borough with the highese total price increase from 1998 to 2018
Total_Increase_DF.plot(x= "Borough Name", y= "Total Price Increase", kind="bar")
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.legend(["Total 20 Year Price Increase"])
plt.show()
```

```python
# Pivot Avg_PPY_DF to have years as columns
pivot_df = Avg_PPY_DF.pivot(index='Borough Name', columns='Year', values='Average Price')

# Calculate the ratio for each borough
pivot_df['Ratio'] = pivot_df[2018] / pivot_df[1998]
```

```python
#Sort by highest ratio and reset index
sorted_AVG_df = pivot_df.sort_values(by='Ratio', ascending=False)
sorted_AVG_df = sorted_AVG_df.reset_index(drop=False)
sorted_AVG_df.head()
```

```python
#Bar plot of the boroughs average increase sorted from greatest to lowest
sorted_AVG_df.plot(x= "Borough Name" , y="Ratio", kind="bar")
```
